```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Make a request
messages = client.messages.create(
    model=model,
    messages=[
        {
            "role": "user", 
            "content": "What is quantum computing? Answer in one sentences"}
    ],
    max_tokens=1000,
)
```

```python
messages.content[0].text
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

# Make a request
def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python
# Make string list of messages
messages = []

add_user_message(messages, "What is quantum computing? Answer in one sentences")
answer = chat(messages)

add_assistant_message(messages, answer)

add_user_message(messages, "Write another senctence.")

answer = chat(messages)
answer
```
